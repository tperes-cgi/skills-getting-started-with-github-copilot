```python
# Créer une liste de produits avec trois éléments
produits = ['tele', 'casque', 'video']

# Afficher la liste complète
print(produits)

# Afficher le premier élément de la liste (index 0)
print(produits[0])

# Modifier le deuxième élément (index 1) de la liste
produits[1] = 'ecouteur'

# Afficher la liste modifiée
print(produits)
```
